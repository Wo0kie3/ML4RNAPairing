```python
import pandas as pd
import xgboost as xgb
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
import warnings
warnings.filterwarnings('ignore') 
```

```python
data = pd.read_csv("full_data_50.csv")
data.head()
```

```python
X = data.drop(['LW', 'Can', 'Saenger'], axis=1)
y_canonical = data['Can']
y_saenger = data['Saenger']
y_lw = data['LW'] # cannonical
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
y_labeled = label_encoder.fit_transform(y_lw)

X_train, X_test, y_train, y_test = train_test_split(X, y_labeled, test_size=0.2, random_state=42)

rus = RandomUnderSampler(random_state=0, sampling_strategy='majority')
X_resampled, y_resampled = rus.fit_resample(X_train, y_train)

model = xgb.XGBClassifier()

model.fit(X_train, y_train)

predictions = model.predict(X_test)

y_test = label_encoder.inverse_transform(y_test)
predictions = label_encoder.inverse_transform(predictions)

accuracy = accuracy_score(y_test, predictions)

print("Accuracy:", accuracy)
print("\nClassification Report:")
print(classification_report(y_test, predictions))
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
y_labeled = label_encoder.fit_transform(y_lw)

X_train, X_test, y_train, y_test = train_test_split(X, y_labeled, test_size=0.1, random_state=42)

rus = RandomOverSampler(random_state=0, sampling_strategy='minority')
X_train, y_train = rus.fit_resample(X_train, y_train)


model = xgb.XGBClassifier()

model.fit(X_train, y_train)

predictions = model.predict(X_test)

y_test = label_encoder.inverse_transform(y_test)
predictions = label_encoder.inverse_transform(predictions)

accuracy = accuracy_score(y_test, predictions)

print("Accuracy:", accuracy)
print("\nClassification Report:")
print(classification_report(y_test, predictions))
```
