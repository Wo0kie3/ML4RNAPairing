```python
import pandas as pd
import xgboost as xgb
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
import warnings
warnings.filterwarnings('ignore') 
```

```python
base = pd.read_csv("stacked_full.csv")
base["base"] = 1
base.head()
```

```python
non_base = pd.read_csv("non_base_some.csv")
non_base.head()
```

```python
X_base = base.drop([str(i) for i in range(50,150)], axis=1)
X_base = X_base.drop(['lw', 'can', 'sae', 'base'], axis=1)
```

```python
X_base.head()
```

```python
X_non_base = non_base.drop(['base'], axis=1)
```

```python
X_non_base.head()
```

```python
X = pd.concat([X_base, X_non_base], ignore_index=True, sort=False)
y_non_base = non_base["base"]
y_base = base["base"]
y = pd.concat([y_base, y_non_base], ignore_index=True, sort=False)
```

# Base

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
y_labeled = label_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_labeled, test_size=0.2, random_state=42)

model = xgb.XGBClassifier()

model.fit(X_train, y_train)

predictions = model.predict(X_test)

y_test = label_encoder.inverse_transform(y_test)
predictions = label_encoder.inverse_transform(predictions)

accuracy = accuracy_score(y_test, predictions)

print("Accuracy:", accuracy)
print("\nClassification Report:")
print(classification_report(y_test, predictions))
```
