```python
import pandas as pd
import xgboost as xgb
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore') 
```

```python
data = pd.read_csv("data_04042024.csv")
data.head()
```

```python
data["sae"].unique()
```

```python
data["lw"].unique()
```

```python
X_lwsae = data[(data.can == False)]
X_lwsae
```

```python
X_LwSae = data[(data.can == False)].drop(['lw', 'can', 'sae'], axis=1)
y_Lw = data[(data.can == False)]["lw"]
y_Sae = data[(data.can == False)]["sae"]
```

```python
X_can = data.drop(['lw', 'can', 'sae'], axis=1)
y_can = data['can']
```

# Canonical

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
y_labeled = label_encoder.fit_transform(y_can)

X_train, X_test, y_train, y_test = train_test_split(X_can, y_labeled, test_size=0.2, random_state=42)
print(len(X_train),len(X_test))
model = xgb.XGBClassifier()

model.fit(X_train, y_train)

predictions = model.predict(X_test)

y_test = label_encoder.inverse_transform(y_test)
predictions = label_encoder.inverse_transform(predictions)

accuracy = accuracy_score(y_test, predictions)

print("Accuracy:", accuracy)
print("\nClassification Report:")
print(classification_report(y_test, predictions))
```
